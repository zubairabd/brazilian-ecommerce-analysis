```python
# 1. Install library
!pip install kagglehub

# 2. Import library
import kagglehub
import pandas as pd
import os

# 3. Download dataset
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")
print("Path to dataset files:", path)

# 4. Cek daftar file untuk memastikan nama file benar
print("Daftar file:", os.listdir(path))
```

```python
# Membaca data menggunakan path yang sudah didefinisikan sebelumnya
df_orders = pd.read_csv(f"{path}/olist_orders_dataset.csv")
df_customers = pd.read_csv(f"{path}/olist_customers_dataset.csv")

# Menampilkan 5 data teratas
df_orders.head()
```

```python
import plotly.express as px

status_counts = df_orders['order_status'].value_counts().reset_index()
fig = px.bar(status_counts, x='order_status', y='count', title='Distribusi Status Pesanan')
fig.show()
```

```python
#load tabel pendukung
df_items = pd.read_csv(f"{path}/olist_order_items_dataset.csv")
df_customers = pd.read_csv(f"{path}/olist_customers_dataset.csv")

# proses merging
#gabung orders dengan items
master_data = pd.merge(df_orders, df_items, on='order_id', how='inner')
#gabung customers
master_data = pd.merge(master_data, df_customers, on='customer_id', how='inner')

#ubah kolom tanggal menjadi format yang sesuai
master_data['order_purchase_timestamp'] = pd.to_datetime(master_data['order_purchase_timestamp'])

print("ukuran data gabungan:", master_data.shape)
master_data.head()

```

```python
import plotly.express as px

# Ekstrak Bulan & Tahun
master_data['month_year'] = master_data['order_purchase_timestamp'].dt.to_period('M').astype(str)

# Hitung total revenue per bulan
monthly_sales = master_data.groupby('month_year')['price'].sum().reset_index()

# Visualisasi interaktif
fig = px.line(monthly_sales, x='month_year', y='price',
              title='Tren Penjualan Bulanan (Olist)',
              markers=True)
fig.update_layout(xaxis_title='Bulan', yaxis_title='Total Pendapatan (BRL)')
fig.show()
```

```python
# Hitung jumlah pesanan per negara bagian
state_sales = master_data.groupby('customer_state')['order_id'].nunique().reset_index().sort_values(by='order_id', ascending=False)

fig_map = px.bar(state_sales, x='customer_state', y='order_id',
                 title='Distribusi Pesanan Berdasarkan Negara Bagian',
                 color='order_id')
fig_map.show()
```

```python
# Load data geolocation
geo = pd.read_csv(f"{path}/olist_geolocation_dataset.csv")

# Ambil rata-rata koordinat per zip code (agar data lebih ringkas)
geo_clean = geo.groupby('geolocation_zip_code_prefix').agg({
    'geolocation_lat': 'mean',
    'geolocation_lng': 'mean'
}).reset_index()

# Gabungkan dengan data master
# master_data 'customer_zip_code_prefix'
map_data = pd.merge(
    master_data,
    geo_clean,
    left_on='customer_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='inner'
)
```

```python
# Tentukan tanggal referensi (asumsikan hari ini adalah 1 hari setelah transaksi terakhir di dataset)
snapshot_date = master_data['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

# Menghitung Recency, Frequency, dan Monetary per pelanggan
rfm = master_data.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days, # Recency
    'order_id': 'count',                                                # Frequency
    'price': 'sum'                                                       # Monetary
})

# Ganti nama kolom agar jelas
rfm.columns = ['Recency', 'Frequency', 'Monetary']

# Tampilkan 5 data teratas
rfm.head()
```

```python
# Memberikan skor 1-4 (makin kecil Recency makin bagus, makin besar F & M makin bagus)
rfm['R_score'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1])
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4])
rfm['M_score'] = pd.qcut(rfm['Monetary'], 4, labels=[1, 2, 3, 4])

# Gabungkan skor
rfm['RFM_Group'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

# Contoh Pelabelan Sederhana
def segment_me(df):
    if df['RFM_Group'] == '444':
        return 'Best Customers (Champions)'
    elif df['R_score'] == 4:
        return 'New Customers'
    elif df['R_score'] == 1:
        return 'Lost Customers / At Risk'
    else:
        return 'Regular'

rfm['Segment'] = rfm.apply(segment_me, axis=1)
print(rfm['Segment'].value_counts())
```

```python
import plotly.express as px

# 1. Hitung jumlah pelanggan per segmen
segment_counts = rfm['Segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'Customer_Count']

# 2. Visualisasi Treemap
fig = px.treemap(segment_counts,
                 path=['Segment'],
                 values='Customer_Count',
                 title='Proporsi Segmen Pelanggan Berdasarkan RFM',
                 color='Customer_Count',
                 color_continuous_scale='RdYlGn')

fig.show()
```
